## Task 1: Python Basics and Operators


```python
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")

if b != 0:
    print(f"{a} / {b} = {a / b}")
    print(f"{a} % {b} = {a % b}")
else:
    print("Division and modulus by zero are not defined.")

print(f"{a} ** {b} = {a ** b}")

print(f"{a} > {b}  -> {a > b}")
print(f"{a} < {b}  -> {a < b}")
print(f"{a} >= {b} -> {a >= b}")
print(f"{a} <= {b} -> {a <= b}")
print(f"{a} == {b} -> {a == b}")
print(f"{a} != {b} -> {a != b}")

a_positive = a > 0
b_positive = b > 0

print(f"a > 0 and b > 0 -> {a_positive and b_positive}")
print(f"a > 0 or b > 0  -> {a_positive or b_positive}")
print(f"not (a > 0)     -> {not a_positive}")
```

## Task 2: Lists and Arrays


```python
import random
import numpy as np

numbers = [random.randint(1, 100) for _ in range(10)]
print(f"Original list: {numbers}")

numbers.append(random.randint(1, 100))
print(f"After adding an element: {numbers}")

if numbers:
    removed = numbers.pop(0)
    print(f"Removed first element ({removed}): {numbers}")

maximum = max(numbers)
minimum = min(numbers)
print(f"Maximum value: {maximum}")
print(f"Minimum value: {minimum}")

sorted_numbers = sorted(numbers)
print(f"Sorted list: {sorted_numbers}")

array = np.array(sorted_numbers, dtype=float)
mean = np.mean(array)
median = np.median(array)
std_dev = np.std(array)

print(f"NumPy array: {array}")
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {std_dev}")
```

## Task 3: Dictionaries and Sets


```python
marks = float(input("Enter student marks (0â€“100): "))

student = {
    "name": input("Enter student name: "),
    "course": input("Enter course name: "),
    "marks": marks,
}

if marks >= 90:
    grade = "A"
elif marks >= 75:
    grade = "B"
elif marks >= 60:
    grade = "C"
else:
    grade = "D"

student["grade"] = grade

print("Student dictionary:")
for key, value in student.items():
    print(f"{key}: {value}")

ai_tools_set1 = {"TensorFlow", "PyTorch", "Keras", "NumPy"}
ai_tools_set2 = {"PyTorch", "Scikit-learn", "Pandas", "NumPy"}

union_tools = ai_tools_set1.union(ai_tools_set2)
intersection_tools = ai_tools_set1.intersection(ai_tools_set2)
difference_tools = ai_tools_set1.difference(ai_tools_set2)

print(f"AI tools set 1: {ai_tools_set1}")
print(f"AI tools set 2: {ai_tools_set2}")
print(f"Union: {union_tools}")
print(f"Intersection: {intersection_tools}")
print(f"Difference (set1 - set2): {difference_tools}")
```

## Task 4: File Handling


```python
def compute_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 75:
        return "B"
    elif marks >= 60:
        return "C"
    else:
        return "D"


students = [
    {"name": "Alice", "marks": 82},
    {"name": "Bob", "marks": 68},
    {"name": "Charlie", "marks": 91},
    {"name": "Diana", "marks": 77},
    {"name": "Ethan", "marks": 54},
]

for student in students:
    student["grade"] = compute_grade(student["marks"])

filename = "ai_students.txt"

with open(filename, "w", encoding="utf-8") as file:
    for student in students:
        line = f"{student['name']},{student['marks']},{student['grade']}\n"
        file.write(line)

print(f"Student data written to {filename}")

print("Students who scored above 75 marks:")
with open(filename, "r", encoding="utf-8") as file:
    for line in file:
        name, marks_str, grade = line.strip().split(",")
        marks = float(marks_str)
        if marks > 75:
            print(f"{name} - Marks: {marks}, Grade: {grade}")
```
